# Animate Line plot from March Man-ness 2 Data


```python
import pandas as pd
```

```python
df = pd.read_csv('./csvData/nicheChefTest.csv')
df
```

```python
df.dtypes
```

```python
df.rename(columns={"A": "a", "B": "c"})
```

```python
def tedCount(value):
    instances = 0
    if value == "Ted [Fueled By Insanity]": instances += 1
    return instances
 
df['ted_count'] = df["Unnamed: 1"].map(tedCount)
display(df.head())
```

```python
match1 = df["Unnamed: 1"]
match1
```

```python
tedList = []
tedCount = 0
for index, value in match1.items():
    if(value == "Ted [Fueled By Insanity]"):
        tedCount+=1
    tedList.append(tedCount)
    
print(tedList)
```

```python
aidenList = []
aidenCount = 0
for index, value in match1.items():
    if(value == "Aiden [Lust Shards]"):
        aidenCount+=1
    aidenList.append(aidenCount)
    
print(aidenList)
```

```python
import matplotlib.pyplot as plt
```

```python
x = []
for i in range(64):
    x.append(i+1)
print(x)
```

```python
plt.plot(x, tedList, label = "Ted")
plt.plot(x, aidenList, label = "Aiden")
plt.legend()
plt.show()
```

```python
import matplotlib
from matplotlib import animation
from matplotlib.animation import FuncAnimation
%matplotlib qt
```

```python
upDownT, upDownA, leftRight = [],[],[]
fig = plt.figure(figsize=(10,8))
axes = fig.add_subplot(1,1,1)
axes.set_ylim(0, 63)
axes.set_xlim(0, 63)
plt.style.use("seaborn")
```

```python

def animate(i):
    if(i ==0): plt.legend()
    
    upDownT.append(tedList[i])
    upDownA.append(aidenList[i])
    leftRight.append(x[i])
    plt.plot(leftRight,upDownT, scaley=True, scalex=True,color = "blue", label = "Ted")
    plt.plot(leftRight,upDownA, scaley=True, scalex=True,color="red", label = "Aiden")
    if(i >= 63): 
        plt.close()
        return
    
    
ani = FuncAnimation(fig=fig, func=animate, interval=100)
#ani.show()
```

```python
f = "nicheChef1.gif" 
writergif = animation.PillowWriter(fps=30) 
ani.save(f, writer=writergif)
#Oh my goodness it still works despite the errors
```

```python
import random
import matplotlib
from matplotlib import animation
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from itertools import count
%matplotlib qt

l1 = [random.randint(-10,4)+(i**1.68)/(random.randint(13,14)) for i in range(0,160,2)]
l2 = [random.randint(0,4)+(i**1.5)/(random.randint(9,11)) for i in range(0,160,2)]
l3 = [random.randint(-10,10)-(i**1.3)/(random.randint(11,12)) for i in range(0,160,2)]
l4 = [random.randint(0,4)+(i**1.6)/(random.randint(10,13)) for i in range(0,160,2)]

fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (15,5))
axes.set_ylim(-100, 500)
axes.set_xlim(0, 250)
plt.style.use("ggplot")

x1,y1,y2,y3 = [], [], [], []
xval = count(0,3)
def animate(i):
    x1.append(next(xval))
    y1.append((l1[i]))
    y2.append((l2[i]))
    y3.append((l3[i]))

    axes.plot(x1,y1, color="red")
    axes.plot(x1,y2, color="gray", linewidth=0.5)
    axes.plot(x1,y3, color="blue")
    
anim = FuncAnimation(fig, animate, interval=30)
```

```python
from matplotlib import pyplot as plt 
from matplotlib import animation

fig = plt.figure()

ax = plt.axes(xlim=(0, 2), ylim=(0, 100))

N = 3
lines = [plt.plot([], [])[0] for _ in range(N)]


def init():    
    for line in lines:
        line.set_data([], [])
    return lines

def animate(i):
    for j,line in enumerate(lines):
        line.set_data([0, 2], [10*j,i])
        line.set_label('line %d, stage %d'%(j,i))
    return lines

anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=100, interval=20, blit=True)
```
